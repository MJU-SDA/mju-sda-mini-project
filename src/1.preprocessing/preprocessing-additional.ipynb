### **2. CCTV 개수 데이터 지역별 정리**

```python
##### 수정

# ------------------------------------------------
# 공통: 시도 이름 정규화 함수 (강원특별자치도 → 강원도 등)
# ------------------------------------------------
target_values = [
    '전국', '서울특별시', '부산광역시', '대구광역시', '인천광역시',
    '광주광역시', '대전광역시', '울산광역시',
    '세종특별자치시',
    '경기도',
    '강원도', '충청북도',
    '충청남도', '전라북도', '전라남도',
    '경상북도', '경상남도', '제주특별자치도'
]

# 새 이름 → 우리가 쓰는 표준 이름
sido_alias = {
    "강원특별자치도": "강원도",
    "전북특별자치도": "전라북도",
}

# 앞글자 기준으로 잡는 prefix
sido_prefix = {
    "서울": "서울특별시",
    "부산": "부산광역시",
    "대구": "대구광역시",
    "인천": "인천광역시",
    "광주": "광주광역시",
    "대전": "대전광역시",
    "울산": "울산광역시",
    "세종": "세종특별자치시",
    "경기": "경기도",
    "강원": "강원도",
    "충북": "충청북도",
    "충남": "충청남도",
    "전북": "전라북도",
    "전남": "전라남도",
    "경북": "경상북도",
    "경남": "경상남도",
    "제주": "제주특별자치도",
}
```

```python
def normalize_sido(raw):
    raw = str(raw).replace(" ", "").strip()

    # 1) alias
    for alias, canon in sido_alias.items():
        if alias.replace(" ", "") in raw:
            return canon

    # 2) 이미 우리가 쓰는 이름이면
    for city in target_values:
        if city.replace(" ", "") == raw:
            return city

    # 3) 접두어 매핑
    for key, val in sido_prefix.items():
        if raw.startswith(key):
            return val

    return raw
```

```python
# wide → long (연번/시도/시군구/CCTV 세트 3개를 세로로)
blocks = []
for suffix in ["", ".1", ".2"]:
    cols = [f"연번{suffix}", f"시도{suffix}", f"시군구{suffix}", f"CCTV{suffix}"]
    sub = cctv_raw[cols].copy()
    sub.columns = ["연번", "시도", "시군구", "CCTV"]
    blocks.append(sub)

c_long = pd.concat(blocks, ignore_index=True)

# 시도 결측치 채우고, 시군구 / CCTV 없는 행 제거
c_long["시도"] = c_long["시도"].ffill()
c_long = c_long.dropna(subset=["시군구", "CCTV"])

# 문자열 정리
c_long["시군구"] = (
    c_long["시군구"].astype(str)
    .str.replace("\n", "", regex=False)
    .str.replace(" ", "", regex=False)
    .str.strip()
)
c_long["시도"] = c_long["시도"].astype(str).str.strip()

# CCTV 숫자형
c_long["CCTV"] = (
    c_long["CCTV"].astype(str)
    .str.replace(",", "", regex=False)
    .str.strip()
)
c_long = c_long[c_long["CCTV"] != ""]
c_long["CCTV"] = c_long["CCTV"].astype(int)

# ============================================
# 2. 행정구역(시) / 행정구역(시군구)별 생성
#    - "시통합관제", "도통합관제" 처럼 '통합'이 들어간 행은
#      → 해당 시 전체 합계로 취급
#      → 행정구역(시군구)별 = 행정구역(시)
# ============================================
c_long["행정구역(시)"] = c_long["시도"].apply(normalize_sido)
c_long["행정구역(시군구)별"] = c_long["시군구"]

# '통합'이 들어가면 시 전체로 매핑 (광주 시 통합관제, 대전 시 통합관제, 제주 도 통합관제 등)
mask_integration = c_long["시군구"].str.contains("통합", regex=False)
c_long.loc[mask_integration, "행정구역(시군구)별"] = c_long.loc[mask_integration, "행정구역(시)"]

# 시청만 따로 있는 행은 버리기 (단, '통합'이 같이 들어간 건 이미 위에서 처리)
mask_drop = c_long["시군구"].str.contains("시청", regex=False) & ~mask_integration
c_long = c_long[~mask_drop].copy()

# ============================================
# 3. 시·군·구별 CCTV 집계
# ============================================
cctv_grouped = (
    c_long
    .groupby(["행정구역(시)", "행정구역(시군구)별"], as_index=False)["CCTV"]
    .sum()
)

# ============================================
# 4. 시 전체 합계 행 추가 (없는 시만 추가)
# ============================================

# (1) 이미 "시 전체 행"이 있는 경우 찾아놓기
#     예: 서울 / 서울특별시
existing_city_keys = {
    (row["행정구역(시)"], row["행정구역(시군구)별"])
    for _, row in cctv_grouped.iterrows()
    if row["행정구역(시)"] == row["행정구역(시군구)별"]
}

# (2) 시군구별 데이터를 시 기준으로 합계 만들기
city_totals = (
    cctv_grouped
    .groupby("행정구역(시)", as_index=False)["CCTV"]
    .sum()
)

# (3) 시전체 행 키 만들기
city_totals["행정구역(시군구)별"] = city_totals["행정구역(시)"]

# (4) 튜플 키 컬럼 생성 (비교용)
city_totals["key"] = list(zip(city_totals["행정구역(시)"], city_totals["행정구역(시군구)별"]))

# (5) 이미 존재하는 (시, 시) 키는 제외하고 추가할 시만 남기기
city_totals = city_totals[~city_totals["key"].isin(existing_city_keys)].copy()

# (6) key 컬럼 제거
city_totals = city_totals.drop(columns=["key"])

# (7) city_totals를 원본과 병합
cctv_df = pd.concat([cctv_grouped, city_totals], ignore_index=True)

# ============================================
# 5. 정렬 (전국 → 서울특별시 → … → 제주특별자치도, 각 시 안에서 합계 먼저)
# ============================================
city_order_map = {city: i for i, city in enumerate(target_values)}
cctv_df["시_순서"] = cctv_df["행정구역(시)"].map(city_order_map).fillna(len(target_values))
cctv_df["is_total"] = (cctv_df["행정구역(시)"] == cctv_df["행정구역(시군구)별"]).astype(int)

cctv_df = cctv_df.sort_values(
    by=["시_순서", "is_total", "행정구역(시군구)별"],
    ascending=[True, False, True]
).reset_index(drop=True)

cctv_df = cctv_df.drop(columns=["시_순서", "is_total"])

print(cctv_df[cctv_df["행정구역(시)"].isin(["광주광역시", "대전광역시", "제주특별자치도"])])
print(cctv_df.shape)
```

### **3. 1인가구 비율 데이터 지역 별 정리**

```python
# -----------------------------------------------------------
# 1) 원본 정리
# -----------------------------------------------------------
one_raw = one_raw[['행정구역별', '2023']].copy()
one_raw = one_raw.rename(columns={'2023': '1인가구비율'})

# -----------------------------------------------------------
# 2) 행정구역(시) 생성(시도 표준화 적용)
# -----------------------------------------------------------
one_raw["행정구역(시)"] = None
current_city = None

for idx, raw in one_raw["행정구역별"].items():
    std = normalize_sido(raw)         # 강원특별자치도 → 강원도 등
    if std in target_values:          # 표준 시도명이라면
        current_city = std
    one_raw.at[idx, "행정구역(시)"] = current_city

# -----------------------------------------------------------
# 3) 행정구역(시군구)별 생성
# -----------------------------------------------------------
one_raw["행정구역(시군구)별"] = (
    one_raw["행정구역별"]
    .astype(str)
    .str.replace(" ", "", regex=False)
)

# -----------------------------------------------------------
# 4) 숫자 변환
# -----------------------------------------------------------
one_raw['1인가구비율'] = (
    one_raw['1인가구비율']
    .astype(str)
    .str.replace(",", "", regex=False)
    .str.replace(r"[^0-9\.]", "", regex=True)
    .str.strip()
)

one_raw = one_raw[one_raw['1인가구비율'] != ""]
one_raw['1인가구비율'] = one_raw['1인가구비율'].astype(float)

# -----------------------------------------------------------
# 5) 최종 정리
# -----------------------------------------------------------
one_person_df = one_raw[['행정구역(시)', '행정구역(시군구)별', '1인가구비율']].copy()
one_person_df = one_person_df.iloc[1:].reset_index(drop=True)

print(one_person_df.head())
print(one_person_df.shape)
```

### **4. 경찰서 수 데이터 지역 별 정리**

```python
target_values = [
    '전국', '서울특별시', '부산광역시', '대구광역시', '인천광역시',
    '광주광역시', '대전광역시', '울산광역시',
    '세종특별자치시', '경기도', '강원도', '충청북도', '충청남도',
    '전라북도', '전라남도', '경상북도', '경상남도', '제주특별자치도'
]

def extract_sido_from_loc(loc):
    """위치 컬럼에서 시/도 이름 추출 (예: '서울특별시청' -> '서울특별시')"""
    loc = str(loc).strip()
    for s in target_values:
        if s.replace(" ", "") in loc.replace(" ", ""):
            return s
    # 그래도 못 찾으면 공백 기준 첫 토큰 (백업 플랜)
    return loc.split()[0] if loc else loc

police_raw["행정구역(시)"] = police_raw["위치"].apply(extract_sido_from_loc)

# -----------------------------
# 2. 행정구역(시군구)별 만들기: '경찰서주소'에서 구/군/시 추출
# -----------------------------
# 예:
#  - "중구 수표로 27"        -> "중구"
#  - "강 남 구 테헤란로..."   -> "강남구"
#  - "제주시 연북로 123"     -> "제주시"
addr_clean = (
    police_raw["경찰서주소"]
    .astype(str)
    .str.replace(" ", "", regex=False)   # "강 남 구" -> "강남구"
)

sgg = addr_clean.str.extract(r'^(?P<sgg>.+?[구군시])')["sgg"]
police_raw["행정구역(시군구)별"] = sgg

# -----------------------------
# 3. 결측/이상치 제거 후 .copy()로 경고 방지
# -----------------------------
police_raw = police_raw.dropna(subset=["행정구역(시)", "행정구역(시군구)별"]).copy()

# 공백 제거
police_raw["행정구역(시)"] = police_raw["행정구역(시)"].str.replace(" ", "", regex=False)
police_raw["행정구역(시군구)별"] = police_raw["행정구역(시군구)별"].str.replace(" ", "", regex=False)

# -----------------------------
# 4. 시·구별 경찰서 개수 집계
# -----------------------------
police_count = (
    police_raw
    .groupby(["행정구역(시)", "행정구역(시군구)별"], as_index=False)
    .size()
    .rename(columns={"size": "경찰서수"})
)

# -----------------------------
# 5. 행정구역(시) 이름을 target_values 기준으로 자동 매핑
#    (강원특별자치도 -> 강원도, 전북특별자치도 -> 전라북도 등)
# -----------------------------
def map_to_target(city_name):
    city_name = str(city_name)
    # 1) target_values에 포함된 이름이 부분 일치하면 그걸 사용
    for t in target_values:
        if t.replace(" ", "") in city_name.replace(" ", ""):
            return t
    # 2) 앞 두 글자로 시작하는 것 찾아보기 (충북 -> 충청북도)
    short = city_name[:2]
    for t in target_values:
        if t.startswith(short):
            return t
    # 3) 그래도 못 찾으면 원래 값 반환
    return city_name

police_count["행정구역(시)"] = police_count["행정구역(시)"].apply(map_to_target)

# 최종 결과 확인
print(police_count.head())
print(police_count.shape)
```

```python
# -----------------------------
# ⭐ 행정구역(시) 전체 합계 행 추가
#   (서울특별시 / 서울특별시, 부산광역시 / 부산광역시 …)
# -----------------------------

# 1) 시 단위로 경찰서수 합계 구하기
city_totals = (
    police_count
    .groupby("행정구역(시)", as_index=False)["경찰서수"]
    .sum()
)

# 2) 시 전체 행은 "행정구역(시군구)별" 도 시 이름으로 맞춰주기
city_totals["행정구역(시군구)별"] = city_totals["행정구역(시)"]

# 3) 원래 시군구 단위 행 + 시 전체 합계 행 합치기
police_count = pd.concat([police_count, city_totals], ignore_index=True)

# 4) 혹시 같은 키가 중복됐으면 하나만 남기기 (안전용)
police_count = (
    police_count
    .drop_duplicates(subset=["행정구역(시)", "행정구역(시군구)별"], keep="first")
    .reset_index(drop=True)
)
```
