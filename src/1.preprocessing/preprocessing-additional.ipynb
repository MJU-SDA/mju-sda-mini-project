### **2. CCTV 개수 데이터 지역별 정리**

```python
# 1) wide → long 변환 (연번/시도/시군구/CCTV 세트 3개를 세로로 붙이기)
blocks = []
for suffix in ["", ".1", ".2"]:
    cols = [f"연번{suffix}", f"시도{suffix}", f"시군구{suffix}", f"CCTV{suffix}"]
    sub = cctv_raw[cols].copy()
    sub.columns = ["연번", "시도", "시군구", "CCTV"]
    blocks.append(sub)

cctv_long = pd.concat(blocks, ignore_index=True)

# 2) 시도 결측치 채우기 + 필요 행만 남기기
cctv_long["시도"] = cctv_long["시도"].ffill()
cctv_long = cctv_long.dropna(subset=["시군구", "CCTV"])

# 3) 제주 "도 통합관제" 행에서 CCTV 합계만 따로 뽑아두기 (나중에 전체 제주에 쓰기)
jeju_mask = (
    (cctv_long["시도"] == "제주")
    & cctv_long["시군구"].astype(str).str.contains("통합", regex=False)
)
jeju_total = (
    cctv_long.loc[jeju_mask, "CCTV"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .astype(int)
    .sum()
)

# 4) 문자열 정리 + 숫자형 변환
c = cctv_long.copy()

c["시군구"] = (
    c["시군구"].astype(str)
    .str.replace("\n", "", regex=False)
    .str.replace(" ", "", regex=False)
    .str.strip()
)
c["시도"] = c["시도"].astype(str).str.strip()

c["CCTV"] = (
    c["CCTV"].astype(str)
    .str.replace(",", "", regex=False)
    .str.strip()
)
c = c[c["CCTV"] != ""]
c["CCTV"] = c["CCTV"].astype(int)

# 5) 시청/통합관제 행들은 시군구 단위 분석에서 애매하니까 제거
#    (제주는 이미 3번에서 값 빼놨으니 같이 지워져도 괜찮음)
mask_drop = c["시군구"].str.contains("통합관제|시청", regex=True)
c = c[~mask_drop].copy()

# 6) 시도 이름을 processed1의 '행정구역(시)' 형식으로 매핑
sido_map = {
    "서울": "서울특별시",
    "부산": "부산광역시",
    "대구": "대구광역시",
    "인천": "인천광역시",
    "광주": "광주광역시",
    "대전": "대전광역시",
    "울산": "울산광역시",
    "세종": "세종특별자치시",
    "경기": "경기도",
    "강원": "강원도",
    "충북": "충청북도",
    "충남": "충청남도",
    "전북": "전라북도",
    "전남": "전라남도",
    "경북": "경상북도",
    "경남": "경상남도",
    "제주": "제주특별자치도",
}

c["행정구역(시)"] = c["시도"].map(sido_map)
c["행정구역(시군구)별"] = c["시군구"]

# 7) 시군구별 CCTV 합계
cctv_grouped = (
    c.groupby(["행정구역(시)", "행정구역(시군구)별"], as_index=False)["CCTV"]
    .sum()
)

# 8) 제주 전체 행 추가 (행정구역(시) = 행정구역(시군구)별 = '제주특별자치도')
jeju_row = pd.DataFrame({
    "행정구역(시)": ["제주특별자치도"],
    "행정구역(시군구)별": ["제주특별자치도"],
    "CCTV": [jeju_total],
})
cctv_df = pd.concat([cctv_grouped, jeju_row], ignore_index=True)

# 9) 시 전체(서울특별시·부산광역시 …) 합계 행 추가
city_totals = cctv_df.groupby("행정구역(시)", as_index=False)["CCTV"].sum()
city_totals["행정구역(시군구)별"] = city_totals["행정구역(시)"]
city_totals = city_totals[["행정구역(시)", "행정구역(시군구)별", "CCTV"]]

#    전국 합계 행도 추가
nation_total = pd.DataFrame({
    "행정구역(시)": ["전국"],
    "행정구역(시군구)별": ["전국"],
    "CCTV": [city_totals["CCTV"].sum()],
})

cctv_df = pd.concat([cctv_df, city_totals, nation_total], ignore_index=True)

# 10) (제주 같은) 완전 동일 키 중복은 하나만 남기기
cctv_df = cctv_df.drop_duplicates(
    subset=["행정구역(시)", "행정구역(시군구)별"],
    keep="first"
).reset_index(drop=True)

# 11) 시/구 정렬 (전에 쓰던 target_values 그대로 사용)
target_values = [
    '전국', '서울특별시', '부산광역시', '대구광역시', '인천광역시',
    '광주광역시', '대전광역시', '울산광역시',
    '세종특별자치시',
    '경기도',
    '강원도', '충청북도',
    '충청남도', '전라북도', '전라남도',
    '경상북도', '경상남도', '제주특별자치도'
]
city_order_map = {city: i for i, city in enumerate(target_values)}

cctv_df["시_순서"] = cctv_df["행정구역(시)"].map(city_order_map).fillna(len(target_values))
cctv_df["is_total"] = (cctv_df["행정구역(시)"] == cctv_df["행정구역(시군구)별"]).astype(int)

cctv_df = cctv_df.sort_values(
    by=["시_순서", "is_total", "행정구역(시군구)별"],
    ascending=[True, False, True]
).reset_index(drop=True)

cctv_df = cctv_df.drop(columns=["시_순서", "is_total"])

# 확인용
print(cctv_df[cctv_df["행정구역(시군구)별"].isin(["서울특별시", "제주특별자치도", "전국"])])
print(cctv_df.shape)
```

### **3. 1인가구 비율 데이터 지역 별 정리**

```python
one_raw = one_raw[['행정구역별', '2023']].copy()
one_raw = one_raw.rename(columns={'2023': '1인가구비율'})


# 3) 시도 리스트 정의
target_values = ['전국', '서울특별시', '부산광역시', '대구광역시', '인천광역시', 
                '광주광역시', '대전광역시', '울산광역시', 
                '세종특별자치시', '경기도', '강원도', '충청북도', '충청남도',
                '전라북도', '전라남도', '경상북도', '경상남도', '제주특별자치도']


# 4) 행정구역(시) 생성 (각 구는 바로 위 행정구역을 상속)
one_raw["행정구역(시)"] = None
current_city = None

for idx, val in one_raw["행정구역별"].items():
    if val in target_values:
        current_city = val
    one_raw.at[idx, "행정구역(시)"] = current_city


# 5) 행정구역(시군구)별 이름 생성
# (시 자체 행은 시 이름 그대로 들어감 → processed1과 merge 시 문제 없음)
one_raw["행정구역(시군구)별"] = one_raw["행정구역별"].str.replace(" ", "")

# 6) 1인가구비율을 float로 변환 (숫자만 남기고 나머지는 제거)
one_raw['1인가구비율'] = (
    one_raw['1인가구비율']
    .astype(str)
    .str.replace(",", "", regex=False)
    .str.replace(r"[^0-9\.]", "", regex=True)   # 숫자와 .만 남기기
    .str.strip()
)

# 빈 문자열(즉 변환 불가능한 텍스트였던 행) 제거
one_raw = one_raw[one_raw['1인가구비율'] != ""]

# float 변환
one_raw['1인가구비율'] = one_raw['1인가구비율'].astype(float)


# 7) 필요한 컬럼만 최종 정리
one_person_df = one_raw[['행정구역(시)', '행정구역(시군구)별', '1인가구비율']].copy()

one_person_df = one_person_df.iloc[1:].reset_index(drop=True)

print(one_person_df.head())
print(one_person_df.shape)
```

```python
area_df = area_df.merge(
    one_person_df,
    on=["행정구역(시)", "행정구역(시군구)별"],
    how="left"
)
area_df
```

### **4. 경찰서 수 데이터 지역 별 정리**

```python
target_values = [
    '전국', '서울특별시', '부산광역시', '대구광역시', '인천광역시',
    '광주광역시', '대전광역시', '울산광역시',
    '세종특별자치시', '경기도', '강원도', '충청북도', '충청남도',
    '전라북도', '전라남도', '경상북도', '경상남도', '제주특별자치도'
]

def extract_sido_from_loc(loc):
    """위치 컬럼에서 시/도 이름 추출 (예: '서울특별시청' -> '서울특별시')"""
    loc = str(loc).strip()
    for s in target_values:
        if s.replace(" ", "") in loc.replace(" ", ""):
            return s
    # 그래도 못 찾으면 공백 기준 첫 토큰 (백업 플랜)
    return loc.split()[0] if loc else loc

police_raw["행정구역(시)"] = police_raw["위치"].apply(extract_sido_from_loc)

# -----------------------------
# 2. 행정구역(시군구)별 만들기: '경찰서주소'에서 구/군/시 추출
# -----------------------------
# 예:
#  - "중구 수표로 27"        -> "중구"
#  - "강 남 구 테헤란로..."   -> "강남구"
#  - "제주시 연북로 123"     -> "제주시"
addr_clean = (
    police_raw["경찰서주소"]
    .astype(str)
    .str.replace(" ", "", regex=False)   # "강 남 구" -> "강남구"
)

sgg = addr_clean.str.extract(r'^(?P<sgg>.+?[구군시])')["sgg"]
police_raw["행정구역(시군구)별"] = sgg

# -----------------------------
# 3. 결측/이상치 제거 후 .copy()로 경고 방지
# -----------------------------
police_raw = police_raw.dropna(subset=["행정구역(시)", "행정구역(시군구)별"]).copy()

# 공백 제거
police_raw["행정구역(시)"] = police_raw["행정구역(시)"].str.replace(" ", "", regex=False)
police_raw["행정구역(시군구)별"] = police_raw["행정구역(시군구)별"].str.replace(" ", "", regex=False)

# -----------------------------
# 4. 시·구별 경찰서 개수 집계
# -----------------------------
police_count = (
    police_raw
    .groupby(["행정구역(시)", "행정구역(시군구)별"], as_index=False)
    .size()
    .rename(columns={"size": "경찰서수"})
)

# -----------------------------
# 5. 행정구역(시) 이름을 target_values 기준으로 자동 매핑
#    (강원특별자치도 -> 강원도, 전북특별자치도 -> 전라북도 등)
# -----------------------------
def map_to_target(city_name):
    city_name = str(city_name)
    # 1) target_values에 포함된 이름이 부분 일치하면 그걸 사용
    for t in target_values:
        if t.replace(" ", "") in city_name.replace(" ", ""):
            return t
    # 2) 앞 두 글자로 시작하는 것 찾아보기 (충북 -> 충청북도)
    short = city_name[:2]
    for t in target_values:
        if t.startswith(short):
            return t
    # 3) 그래도 못 찾으면 원래 값 반환
    return city_name

police_count["행정구역(시)"] = police_count["행정구역(시)"].apply(map_to_target)

# 최종 결과 확인
print(police_count.head())
print(police_count.shape)
```

```python
# -----------------------------
# ⭐ 행정구역(시) 전체 합계 행 추가
#   (서울특별시 / 서울특별시, 부산광역시 / 부산광역시 …)
# -----------------------------

# 1) 시 단위로 경찰서수 합계 구하기
city_totals = (
    police_count
    .groupby("행정구역(시)", as_index=False)["경찰서수"]
    .sum()
)

# 2) 시 전체 행은 "행정구역(시군구)별" 도 시 이름으로 맞춰주기
city_totals["행정구역(시군구)별"] = city_totals["행정구역(시)"]

# 3) 원래 시군구 단위 행 + 시 전체 합계 행 합치기
police_count = pd.concat([police_count, city_totals], ignore_index=True)

# 4) 혹시 같은 키가 중복됐으면 하나만 남기기 (안전용)
police_count = (
    police_count
    .drop_duplicates(subset=["행정구역(시)", "행정구역(시군구)별"], keep="first")
    .reset_index(drop=True)
)
```

```python
area_df = area_df.merge(
    police_count,
    on=["행정구역(시)", "행정구역(시군구)별"],
    how="left"
)
```
